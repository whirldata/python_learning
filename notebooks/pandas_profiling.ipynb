```python
import os
import pandas as pd
import urllib.request
```

```python
directory_to_extract_to = "data"
data_url = "http://eforexcel.com/wp/wp-content/uploads/2017/07/50000-Sales-Records.zip"
```

```python
if not os.path.exists('data'):
    os.makedirs('data')
```

```python
path_to_zip_file = "data/sales.zip"
```

```python
urllib.request.urlretrieve(data_url, path_to_zip_file)
```

```python
import zipfile
with zipfile.ZipFile(path_to_zip_file, 'r') as zip_ref:
    zip_ref.extractall(directory_to_extract_to)
```
